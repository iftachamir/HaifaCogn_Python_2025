## REPL and the Python Interpreter

Python is an interpreted language, which means that code is executed line-by-line by an interpreter. This differs from compiled languages, where the entire program is translated into machine language before execution.

Python provides an interactive environment called the **REPL (Read-Evaluate-Print Loop)**, which executes your commands immediately. This environment is especially useful for experimenting with short snippets of code and for quick testing.

**How the REPL works:**

- **Read**: Python reads the input command you type.
- **Evaluate**: Python evaluates (executes) your command.
- **Print**: Python prints the result or any output produced.
- **Loop**: Python returns to the prompt, awaiting the next command.

You can use this notebook as an interactive environment, similar to a REPL, by running small bits of code in separate cells.

## Errors and Debugging

Encountering errors is a normal and informative part of programming. Errors help you identify and correct issues in your code. Here are the three main types of errors you'll commonly encounter in Python:

### Syntax Errors
Syntax errors occur when Python cannot interpret your code because it violates Python's grammar rules. Python will show an error message and indicate where the problem occurred.


```python
# Syntax error
print('Hello World)
```

### Runtime Errors
Runtime errors occur during the execution of the program. The syntax is correct, but Python encounters an unexpected condition and cannot continue running your code.


```python
# Runtime error example: uncomment and run this code to see the error.
print(1 / 0)  # Division by zero
```

### Debugging - Logic Errors
Logic errors are the most subtle type. Your code runs without any explicit error messages, but it doesn't produce the expected outcome because the underlying logic is incorrect. Meaning Python is "correctly" doing what it's supposed to do - it is the programmer who has overlooked some issue(s).

```python
# Logic error example:
x = 5
print(f'The square of {x} is:', x + x)  # Incorrect logic; should be x * x
```

## Variable Types and Duck Typing

Python is dynamically typed, which means that variables do not have a fixed type. Instead, the type of a variable is determined at runtime based on the value assigned to it. This is sometimes referred to as **duck typing**.

Duck typing follows the principle:

> "If it walks like a duck and quacks like a duck, then it must be a duck."

In practice, Python cares more about the operations that an object supports rather than its explicit type.

**Example:**


```python
# Duck typing example:
print(5 + 10)                # integer addition
print(5.1 + 10)            # float addition
print('Hello' + 'Shalom')  # string concatenation
```

## Giving Descriptive Variable Names

When naming variables, it's best practice to choose descriptive, meaningful names that clearly indicate their purpose or content. Descriptive variable names greatly enhance the readability and maintainability of your code.

### Example of Good vs. Bad Variable Names:

| Poor naming | Descriptive naming |
|-------------|--------------------|
| `a`, `b`, `c`| `age`, `salary`, `temperature`|
| `lst`        | `student_names`|
| `x`, `y`     | `width`, `height`|

When naming variables with multiple words, use underscores `_` to separate the words (this is known as **snake_case**).


## Reserved Names and Existing Function Names

Python has reserved wordsâ€”known as **keywords**â€”that are used to define the languageâ€™s syntax and structure. You **cannot** use these keywords as variable names.

**Examples of Python keywords:**  
`for`, `while`, `if`, `else`, `elif`, `def`, `return`, `True`, `False`, `None`, `and`, `or`, `import`, etc.

Additionally, while Python allows it, it's considered bad practice to use the names of built-in functions (e.g., `list`, `str`, `print`) as variable names because it overrides the original function.

**Example of what NOT to do:**


```python
print(round(4.1))

round = 10
print(round)

print(round(10.1))
```
